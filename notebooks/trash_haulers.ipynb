```python
address_counts = hauler['Incident Address'].value_counts()
```

This is a series where the index is the address, and the values are the number of times each address appears.

```python
repeated_addresses = address_counts[address_counts > 1].index
```

I am filtering the addresses that appear more than once.

```python
hauler = hauler[hauler['Incident Address'].isin(repeated_addresses)]
```

Only the rows where Incident Address is in the repeated_addresses will appear now.

```python
hauler
```

```python
hauler.loc[
    (hauler['Description'].str.contains('miss', regex=True, na=False)) |
    (hauler['Complaint'].str.contains('miss', regex=True, na=False))
].groupby('Incident Address').size()
```

```python
hauler.loc[:, 'Incident Address'] = hauler['Incident Address'].str.split(',').str[0]
```
